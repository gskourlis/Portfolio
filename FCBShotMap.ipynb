```python
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer.pitch import VerticalPitch
```

```python
text_color = 'w'
```

```python
data = pd.read_csv('/Users/georgiosskourlis/Desktop/Football Data/shotmaps.csv')
```

```python
data
```

```python
fig , ax = plt.subplots(figsize=(13,8.5))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = pitch = VerticalPitch(half=True, pitch_color = '#22312b')

pitch.draw(ax=ax)

plt.gca().invert_yaxis()

for x in range(len(data['x'])):
    if data['outcome'][x] == 'Missed':
       plt.scatter(data['x'][x],data['y'][x],color = 'red', s=100)
        
    elif data['outcome'][x] == 'Saved':
       plt.scatter(data['x'][x],data['y'][x],color = 'blue', s=100)
        
    else:
       plt.scatter(data['x'][x],data['y'][x],color = 'yellow', s=100)
        

lb = ['Missed', 'Saved', 'Post']
colors = ['red', 'blue', 'yellow']

legend = plt.legend(fontsize= 15, frameon=False, labels=lb)

for text, color in zip(legend.get_texts(), colors):
    text.set_color(color)
    
plt.title('FC Barcelona Shot Map ', fontsize = 24, c= text_color)

```
